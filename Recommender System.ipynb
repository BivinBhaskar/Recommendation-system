```python
pivot=pd.pivot_table(data=book,index='Userid',columns='Title',values='Rating')
```

```python
pivot
```

```python
pivot.fillna(0,inplace=True)
```

```python
pivot.index=book['Userid'].unique()
pivot
```

```python
## calculate distance and similarities
from sklearn.metrics import pairwise_distances
```

```python
pairwise_distances(pivot,metric='cosine')
```

```python
sim=1-pairwise_distances(pivot,metric='cosine')
sim
```

```python
## treating the diagonal values
np.fill_diagonal(sim,0)
```

```python
simil=pd.DataFrame(sim)
```

```python
simil
```

```python
simil.index=book['Userid'].unique()
simil.columns=book['Userid'].unique()
```
